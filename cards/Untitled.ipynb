```python
from card import Card
from hand import Hand
import csv

with open('poker.txt', 'r') as f:
    hands = list(csv.reader(f, delimiter=' '))
    
player_hands = []
opponent_hands = []

for hand in hands:
    temp = []
    for card in hand[0:5]:
        new_card = Card(*card)
        temp.append(new_card)
        new_card.calculate_card_value()
    player_hands.append(temp)
    
    temp = []
    for card in hand[5:10]:
        new_card = Card(*card)
        new_card.calculate_card_value()
        temp.append(new_card)
    opponent_hands.append(temp)

hands = []

for hand in player_hands:
    hands.append(Hand(hand))
    

for hand in hands:
    hand.hand_type(hand.card_objects)

for hand in hands:
    print(hand.hand_type_one)
```
